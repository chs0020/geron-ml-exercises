### The Perceptron

#### Structure





Linear algebra provides the output for several input instances at once for various **weights** and a **bias** term:

$$ h_{\textbf{W},\textbf{b}}(\textbf{X}) =  \phi (\textbf{X}\textbf{W} + \textbf{b}) $$

- $\textbf{X}$: Matrix of input instances (rows are instances, cols are features)
- $\textbf{W}$: Matrix of weights (one row per input neuron, one col per artificial neuron in the layer)
- $\textbf{b}$: Bias vector (one bias term per artificial neuron)
- $\phi$: The **activation function** (Heaviside step function, sigmoid, etc.)

We can convince ourselves that this will work. If an input has 5 features and 15 instances, there would be 5 input neurons. If there were 3 layers, we'd have a matrix of (15 x 5) * (5 x 3) = (15 x 3). Plus the bias term (one per artificial neuron), we have the column dimension matching.

#### Training

Trained with **Hebb's Rule**. Inputs are fed one at a time. Each neuron that contributed to a wrong prediction, it will reinforce those that would have contributed to a correct prediction:

$$ w_{i,j}^{next step} = w_{i,j} + \eta (y_i - \hat{y}_j)x_i $$

- $ w_{i,j} $: connection weight b/w i'th input neuron and j'th output neuron.
- $ x_i $: i'th input of current training instance (the ith column of whatever row is being fed)
- $ \hat{y}_j $: output of the j'th output neuron given the current training instance
- $ y_j $: target output for the j'th output neuron given the current training instance
- $ \eta $: learning rate

It's similar to _Stochiastic Gradient Descent_ in this regard.

#### Using the Perceptron on the Iris Training Set

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron

iris = load_iris()
X = iris.data[:, (2, 3)]  # petal length, petal width
y = (iris.target == 0).astype(np.int)  # Iris setosa?

per_clf = Perceptron()
per_clf.fit(X, y)

y_pred = per_clf.predict([[2, 0.5]])
print(y_pred)
```

##### Backpropagation

**autodiff**: _Automatic Differentiation_ or _autodiff_ is an algorithm to compute network's gradient w/ regard to each model parameter (weights and biases). Once the gradient is calculated, it steps slightly in that direction. 

Using it w/ backpropagation is called **reverse-mode autodiff**. 

- Typically used with mini-batches (each mini-batch called an _epoch_)
- Instances passed through w/ output of each intermediate result saved for later (when the backwards pass comes)
- Measure the output error w/ a loss function (done this before)
- Measure how much each connection contributed to the error in the output analytically using the **chain rule**.
- Iteratively repeat the last step for each layer in the network. This effectively produces the _error gradient_, or how much the error is changing per output given the inputs fed into it.
- Using all error gradients computed, perform a Gradient step towards a minimum error by adjusting each weight vector in the direction of that minimum (thanks calculus).

Each weight must be initialized randomly to ensure the weights contribute differently to a predicition at first.

Imperative that the loss function has a well-defined gradient everywhere to direct the direction the training should go as the step function's output gradient is flat, not indicating any direction to move. **Nonlinearity is needed because nested linear functions are always linear functions, but nested nonlinear functions may be able to closely approximate any function (think Taylor's Expansion/Approximation).**

##### Regression MLPs

One output neuron per dimension you want to predict. Examples inclue...
- House value would have one output neuron. 
- Car price, gas mileage, and mean lifetime until engine failure would require 3 output neurons
- To place a bounding box around an object in an image would require 2 output neurons for the center of the image and 2 output neurons for the width and the height of the object

Generally, no activation function on the output neurons for regression. For only positive, rectified linear activation function works.

Loss function to use:
- Mean squared error is typical
- A lot of outliers, mean absolute error might work better
- Huber loss is a combo of both (quadratic below a threshold, linear above it to reduce outlier influence)





##### Classification MLPs

One output neuron per classifier. Softplus function typically used on output layer:



Softmax function ensures all output probabilities are between 0 and 1 (which is needed if all output neurons belong to a mutually exclusive class of phenomenon you are trying to predict, like numbers, which means a 9 can't be a 4). 



Loss function to use:
- log-loss/cross entropy loss is typically used

 #### Loading Dataset (Fashion MNIST) with Keras

```python
# Loading Fashion MNIST w/ Keras
fashion_mnist = keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
```

```python
# Structure of the imported data
print("X_train's shape: ", np.shape(X_train_full))
print("X_train's data type: ", X_train_full.dtype)
print("X_train's values: ", X_train_full[0][16])
print("Row of X_train: ", np.shape(X_train_full[0]), "\n")

print("y_train's shape: ", np.shape(y_train_full))
print("y_train's datatype:", y_train_full.dtype)
print("Row of y_train: ", np.shape(y_train_full[0]))
```

#### Shape and Prepare the Data

Must create a validation set before testing against the test set. Each pixel looks like it has a value between 0 and 255. With Gradient Descent, scaling the features to range from 0 to 1 should be done:

```python
# Scaling for Grad Descent and splitting into validation set
X_valid, X_train = X_train_full[:5000] / 255.0, X_train_full[5000:] / 255.0
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

### Creating the Model with the Sequential API

#### Creation and Breakdown of First Model

Classification MPL w/ 2 hidden layers:

```python
# Creating MPL w/ Sequential model
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28, 28]))
model.add(keras.layers.Dense(300, activation="relu"))
model.add(keras.layers.Dense(100, activation="relu"))
model.add(keras.layers.Dense(10, activation="softmax"))
```

**First model breakdown:**

- FIRST LINE: Sequential() = [Keras Sequential model](https://keras.io/models/sequential/). Single stack of layers connected sequentially.
- SECOND LINE: Configures the shape of the instance that will be passed to the model. Receives a data instance (X), and computes X.reshape(-1, 1) like [this](https://docs.scipy.org/doc/numpy/reference/generated/numpy.reshape.html), or brings the 2D image to a 1D input.
     - Alternative: Using an [Input_Layer](https://www.tensorflow.org/api_docs/python/tf/keras/layers/InputLayer) class, configuring the input_shape parameter to be a tuple(28, 28)
- THIRD LINE: Add a [Dense](https://keras.io/layers/core/) _hidden layer_.
     - 1st arg, units=300, dimensionality of output space, or 300 neurons
     - 2nd arg, activation="relu", Rectified Linear activation function
     - Each layer maintains its own weight matrix and bias vector to calculate the output structure $h_{\textbf{W},\textbf{b}}(\textbf{X})$
- FOURTH LINE: Add another Dense hidden layer w/ 100 neurons.
- FIFTH LINE: Add a final Dense _output layer_ with 10 neurons to correspond to the 10 possible classification outputs and using softmax activation to keep all outputs summing to 1, or 100% probability.

Could also add each layer in initialization.

```python
# Adding layers to Sequential model in init
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="relu"),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])
```

#### Visualization and Summary of First Model

We can **visualize the model** and display a **summary of the model** like this: 

```python
# Visualize first model
keras.utils.plot_model(model)
```

```python
model.summary()
```

Can rename layers/set the names of the layers during their creation. "None" in Output Shape means the batch size can be anything. In the first dense layer (dense_9), we are mapping 784 inputs to 300 outputs, which means 784 x 300 = 235200 weight parameters can be adjusted in the weight matrix (one row per input neuron which is 784 for each pixel, and one col per artificial neuron in the layer, which was set to 300). Add onto that 300 bias terms and there are 235500 parameters that can be adjusted in that layer. **kernel** is another name for the **weight matrix**

The **weights and biases** of a dense layer can be accessed by a getter method (or they can be set):

```python
# Weights and biases of dense_9
weights_d9, biases_d9 = model.layers[1].get_weights()
print("Shape of the Weights Matrix of dense_9: ", weights_d9.shape)
print("Shape of the Biases Vector of dense_9: ", biases_d9.shape)
```

#### Compiling the Model

Compile specifies:
   - Optimization method
   - Loss-function
   - Metrics to compute and collect

```python
# Compile the model
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])
```

On the loss function options:

   - "sparse_categorical_crossentropy" for sparse categorization (0 through 9, coming up one per instance). If the labels are in the form of one-hot vectors, they can be converted [using this method](https://docs.scipy.org/doc/numpy/reference/generated/numpy.argmax.html)
   - "categorical_crossentropy" for one-hot [using this method](https://www.tensorflow.org/api_docs/python/tf/keras/utils/to_categorical)
   - "sigmoid" in the last layer and "binary_crossentropy" if we only had a binary output
   
More on optimizing and measuring in later chapters.

#### Training the Model



```python
# Train the model and save the history
history = model.fit(X_train,
                    y_train,
                    epochs=30,
                    validation_data=(X_valid, y_valid))
```

**If performance on the training set is much better on the validation set, there is most likely a significant overfitting going on.**

Instead of passing a validation set, can set the parameter **validation_split** as seen [here](https://keras.rstudio.com/reference/fit.html), which will take the last fraction of whatever training data you pass it and use that as the validation set.

[The history object](https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/History) contains the epochs, training parameters, and metrics collected:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
plt.title("First Keras Model Learning Curve - Fashion MNIST")
plt.xlabel("Training Epoch")
plt.show()
```

 Training curves calculated as a running mean. Validation curves calculated at the end. Therefore, training curves are shifted 1/2 and epoch to the right when they are plotted.
 
 Validation curves are close to the training curves, so it does not look like we are overfitting the model.
 
 Calling .fit() again on this model would continue training this model (and may result in higher accuracy as the loss function does not seem to have converged).

**ON FIXING THE PERFORMANCE OF AN UNDERPERFORMING MODEL:**
- Check the learning rate
- Try a different optimizer
- Tune hyperparameters (more on this later)
 - Number of layers
 - Neurons per layer
 - Activation functions in each hidden layer
 - Batch size

Once performance on the validation set is satisfactory, run on the test set with the [evaluate method](https://keras.io/models/model/#evaluate)

```python
# Evaluating the trained model
model.evaluate(X_test, y_test)
```

### Building a Regression MLP using Sequential API

Use California housing dataset seen in Ch. 2.

- Load
- Split into training, test, and validation sets
- Scale features

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Loading the housing dataset
housing = fetch_california_housing()
```

```python
# Splitting into training, validation, test sets
X_train_full, X_test, y_train_full, y_test = train_test_split(
    housing.data, housing.target)
X_train, X_valid, y_train, y_valid = train_test_split(
    X_train_full, y_train_full)
```

Similar process to classification, but last layer must have only one output neuron (for each dimension being regressed for prediction).

```python
# Define the Sequential model
model = keras.models.Sequential([
    keras.layers.Dense(30, activation="relu", input_shape=X_train.shape[1:]),
    keras.layers.Dense(1)
])
```

```python
# Compile the model
model.compile(loss="mean_squared_error", optimizer="sgd")
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=20,
                    validation_data=(X_valid, y_valid))
```

```python
# Evaluate the model
mse_test = model.evaluate(X_test, y_test)
```

```python
# Make predictions
X_new = X_test[:3] # pretend these are new instances
y_pred = model.predict(X_new)
```

### Complex Models with Functional API

#### Example: Wide and Deep Model (Cheng et al., 2016)



- Simple rules through short path
- Deep patterns through deep path

(Note: underscores are used to prevent conflicting keywords within Python)

```python
# Wide and deep network model
input_ = keras.layers.Input(shape=X_train.shape[1:]) 
hidden1 = keras.layers.Dense(30, activation="relu")(input_)  # Passing input layer as soon as .Dense is created
hidden2 = keras.layers.Dense(30, activation="relu")(hidden1) # Passing hidden1 to indicate the next connection
concat = keras.layers.Concatenate()([input_, hidden2])       # Output of hidden2 and input_ concatenated
output = keras.layers.Dense(1)(concat)                       # Single neuron output (no activation func yet)
model = keras.Model(inputs=[input_], outputs=[output])       # Finalize the model class
```

Passing the layers indicates the connections. Reading more on the [Functional API method of building models](https://keras.io/getting-started/functional-api-guide/)

```python
# Visualize wide and deep
keras.utils.plot_model(model)
```

And what do you know, it looks like our image!

#### Example: Wide and Deep Model (Cheng et al., 2016) with Sectionalized Inputs

- Features 0 thru 4 on Wide path
- Features 2 thru 7 on Deep path

Note the overlap.

```python
# Pass specific inputs for Wide and Deep network
input_A = keras.layers.Input(shape=[5], name="wide_input")
input_B = keras.layers.Input(shape=[6], name="deep_input")
hidden1 = keras.layers.Dense(30, activation="relu")(input_B)
hidden2 = keras.layers.Dense(30, activation="relu")(hidden1)
concat = keras.layers.concatenate([input_A, hidden2])
output = keras.layers.Dense(1, name="output")(concat)
model = keras.Model(inputs=[input_A, input_B], outputs=[output])
```

```python
# Visualize second wide and deep
keras.utils.plot_model(model)
```
